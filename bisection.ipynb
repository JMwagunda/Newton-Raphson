```python
import math
import timeit

def f(x):
    return 2*x**3 - 3**x - 12*x

a = 2
b = -1
tolerance = 1e-6
max_iterations = 100
    
for i in range(max_iterations):
    c = (a + b) / 2
    # print(f"Iteration number is: {i}")
    print("Iterations " + str(i) + ": x = " + str(c) + " f(x) = " +    
          str(f(c)))
    
    if abs(f(c)) < tolerance:
        print(f"Root found at x = {c:6f}")
        print("Time tacken is: ", timeit.timeit())
        break
    elif f(c) * f(a) < 0:
        b = c
    else:
        a = c
```
